# Data Visualization

## Analysing the Total Data

* Total Suicides in Madhya Predesh from 2001 to 2012

```python
years=df["Year"].unique()
year_total=[]
for year in years:
    year_total.append(df['Total'][df['Year']==year].sum())
print(year_total)
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=years,y=year_total)
plt.title("Total suicides from 2001 - 2012 ")
plt.xlabel("Year",fontsize=12,labelpad=15)
plt.ylabel("Total Suicides",fontsize=12,labelpad=15)
```

* Total Suicides in Madhya predesh On the bases of Type Code

```python
type_code=df["Type_code"].unique()
type_code_total=[]
for type_ in type_code:
    type_code_total.append(df['Total'][df['Type_code']==type_].sum())
print(type_code_total)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=type_code,y=type_code_total)
plt.title("Total suicides (Type_Code_wise) ")
plt.xlabel("type_code",fontsize=12,labelpad=15)
plt.ylabel("Total Suicides",fontsize=12,labelpad=15)
```

* Total suicides on the basis of reason of suicide

```python
type_=df['Type'].unique()
type_sum=[]
for x in type_:
    type_sum.append(df["Total"][df["Type"]==x].sum())
print(type_sum)
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x=type_,y=type_sum)
plt.title("Total suicides (Type_wise) ",fontsize=15)
plt.xlabel("Type of Suicides",fontsize=15,labelpad=15)
plt.ylabel("Total Suicides",fontsize=15,labelpad=15)
plt.xticks(rotation=90) 
```

* Total suicides in madhya predesh from 2001 to 2012 Gender Based

```python
gender=df["Gender"].unique()
gender_total=[]
for ge in gender:
    gender_total.append(df["Total"][df["Gender"]==ge].sum())
gender_total
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=gender,y=gender_total)
plt.title("Total suicides (Gender_wise) ")
plt.xlabel("Gender",fontsize=12,labelpad=15)
plt.ylabel("Total Suicides",fontsize=12,labelpad=15) 
```

* Total Suicides according to the Age Group

```python
age_group=df['Age_group'].unique()
age_group_total=[]
for age in age_group:
    age_group_total.append(df['Total'][df['Age_group']==age].sum())
print(age_group_total)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=age_group,y=age_group_total)
plt.title("Total suicides (Age_group_wise) ")
plt.xlabel("Age_Group",fontsize=12,labelpad=15)
plt.ylabel("Total Suicides",fontsize=12,labelpad=15) 
```

* Total suicide in MP from 2001 to 2012 on the basis of gender

```python
years_gender=df["Year"].unique()
year_male_total=[]
year_female_total=[]
for year in years_gender:
    year_male_total.append(df["Total"][(df["Year"]==year) & (df["Gender"]=='Male')].sum())
    year_female_total.append(df["Total"][(df["Year"]==year) & (df["Gender"]=='Female')].sum())
print(year_male_total)
print(year_female_total)
```

```python
plt.figure(figsize=(10,8))
plt.bar(years_gender, year_male_total,0.35)
plt.bar(years_gender+0.35, year_female_total,0.35)
plt.legend(['Male','Female'])
plt.title("Total Suicises of Male and Female(2001-2012)")
plt.xlabel("Year",fontsize=12,labelpad=10)
plt.ylabel("Total_Suicides",fontsize=12,labelpad=10)
```

* Total suicide in MP from 2001 to 2012 on the basis of type code and gender

```python
type_code_gender=df["Type_code"].unique()
type_code_male_total=[]
type_code_female_total=[]
for type_code in type_code_gender:
    type_code_male_total.append(df["Total"][(df['Type_code']==type_code) & (df['Gender']=='Male')].sum())
    type_code_female_total.append(df["Total"][(df['Type_code']==type_code) & (df['Gender']=='Female')].sum())
print(type_code_male_total)
print(type_code_female_total)
```

```python
df_type_code_gender=pd.DataFrame({'type_code':type_code_gender,'male':type_code_male_total,'female':type_code_female_total})
df_type_code_gender.head()
```

```python
ax=df_type_code_gender.plot(kind='bar',x='type_code',y='male',stacked=True,color='teal',alpha=0.6,figsize=(10,8))
df_type_code_gender.plot(kind='bar',x='type_code',y='female',stacked=True,color='orange',alpha=0.5,ax=ax,figsize=(10,8))
plt.title("Total Suicises of Male and Female (Type Code wise)")
plt.xlabel("Type Code",fontsize=12,labelpad=10)
plt.ylabel("Total_Suicides",fontsize=12,labelpad=10)
plt.xticks(rotation=0)
```

* Total suicide in MP from 2001 to 2012 on the basis of reason for Suicide and gender

```python
type_gender=df["Type"].unique()
type_male_total=[]
type_female_total=[]
for type_ in type_gender:
    type_male_total.append(df['Total'][(df['Type']==type_) & (df['Gender']=='Male')].sum())
    type_female_total.append(df['Total'][(df['Type']==type_) & (df['Gender']=='Female')].sum())
    
print(type_male_total)
print(type_female_total)

df_type_gender=pd.DataFrame({"type":type_gender,"male":type_male_total,"female":type_female_total})
```

```python
ax=df_type_gender.plot(kind='bar',x='type',y='male',stacked=True,color='teal',alpha=0.6,figsize=(25,10))
df_type_gender.plot(kind='bar',x='type',y='female',stacked=True,color='orange',alpha=0.5,ax=ax,figsize=(25,10))
plt.title("Total Suicises of Male and Female(Type of Suicide)",fontsize=15)
plt.xlabel("Type",fontsize=15,labelpad=10)
plt.xticks(rotation=90)
plt.ylabel("Total_Suicides",fontsize=15,labelpad=10)
```

* Total suicide in MP from 2001 to 2012 on the basis of age group and gender

```python
age_group_gender=df['Age_group'].unique()
age_group_male_total=[]
age_group_female_total=[]
for age_group in age_group_gender:
    age_group_male_total.append(df['Total'][(df['Age_group']==age_group) & (df['Gender']=='Male')].sum())
    age_group_female_total.append(df['Total'][(df['Age_group']==age_group) & (df['Gender']=='Female')].sum())
print(age_group_male_total)
print(age_group_female_total)
df_age_group_gender=pd.DataFrame({'age_group':age_group_gender,'male':age_group_male_total,'female':age_group_female_total})
df_age_group_gender.head()
```

```python
ax=df_age_group_gender.plot(kind='bar',x='age_group',y='male',stacked=True,color='teal',alpha=0.6,figsize=(10,8))
df_age_group_gender.plot(kind='bar',x='age_group',y='female',stacked=True,color='orange',alpha=0.5,ax=ax,figsize=(10,8))
plt.title("Total Suicises of Male and Female(Age Group wise)")
plt.xlabel("Age_group",fontsize=12,labelpad=10)
plt.ylabel("Total_Suicides",fontsize=12,labelpad=10)
plt.xticks(rotation=0)
```

* Total suicide from 2001-2012 on which age group has what reason for suicide 

```python
type_age_group=df['Type'].unique()
age_14=[]
age_29=[]
age_44=[]
age_59=[]
age_60=[]
age_100=[]
for type_ in type_age_group:
    age_14.append(df['Total'][(df['Type']==type_) & (df['Age_group']=='0-14')].sum())
    age_29.append(df['Total'][(df['Type']==type_) & (df['Age_group']=='15-29')].sum())
    age_44.append(df['Total'][(df['Type']==type_) & (df['Age_group']=='30-44')].sum())
    age_59.append(df['Total'][(df['Type']==type_) & (df['Age_group']=='45-59')].sum())
    age_60.append(df['Total'][(df['Type']==type_) & (df['Age_group']=='60+')].sum())
    age_100.append(df['Total'][(df['Type']==type_) & (df['Age_group']=='0-100+')].sum())

print(age_14)
print(age_29)
print(age_44)
print(age_59)
print(age_60)
print(age_100)
```

```python
df_type_age_group=pd.DataFrame({
    'type':type_age_group,
    '0-14':age_14,
    '15-29':age_29,
    '30-44':age_44,
    '45-59':age_59,
    '60+':age_60,
    '0-100+':age_100
})
df_type_age_group.head()
```

```python
df_type_age_group.set_index("type",inplace=True)
df_type_age_group.head()
```

```python
df_type_age_group.plot(kind='bar',stacked=True,figsize=(25,15))
plt.title('Total suicides by reason of Suicide according to Age Group',fontsize=20)
plt.xlabel("Reason of Suicide",fontsize=15,labelpad=12)
plt.ylabel("Total Number of Suicide",fontsize=15,labelpad=12)
```
